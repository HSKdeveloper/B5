
````python
def generate_password(length: int, lower_alphabet: str, upper_alphabet: str, num: int, min_num: int, symbol: str, symbol_min_num: int):

    pass_char = []

    for n in range (0 , min_num):
        pass_char.append(random.choice(num))

    for n in range (0 , symbol_min_num):
        pass_char.append(random.choice(symbol)) 
    
    remain_length = length-len(pass_char)

    random_pass_char = lower_alphabet + upper_alphabet + num +symbol

    for x in range (0, remain_length):
        pass_char.append(random.choice(random_pass_char)) 

    random.shuffle(pass_char)
    return "".join(pass_char)
````

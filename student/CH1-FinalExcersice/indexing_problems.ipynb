## قائمة: استبدال قيمة عند فهرس معيّن

## قائمة: إيجاد الفهرس بالقيمة ثم الاستبدال

ابحث عن **أول** موضع للقيمة `20` ثم استبدلها بـ `200`:

```python
#| eval: false
values = [10, 20, 30, 20, 40]
```

```python
print(f"values list before change is: {values}")
for num in range(len(values)):
    if values[num] == 20:
        values[num] = 200

print(f"values list after change is: {values}")
```

**المطلوب:** استخدم دالة تبحث بالقيمة (مثل `.index(...)`) ثم الإسناد بالفهرس.



## قائمة: كل عنصر آخر باستخدام `start:stop:step`

من القائمة التالية، استخرج **كل عنصر آخر** بدءًا من الأول، ثم **كل عنصر آخر** بدءًا من الثاني:

```python
#| eval: false
nums = [0, 1, 2, 3, 4, 5, 6, 7]
```

```python
for n in range(0,len(nums)):
    print(nums[n], end=" ")
```

```python
for n in range(7,1,-1):
    print(nums[n], end=" ")
```

**المطلوب:** شريحتان بصيغة `[start:stop:step]` (يمكن إهمال `start`/`stop` عند الحاجة).
